```python
df=pd.read_csv(r"D:\Datasets\auto-mpg.csv")
print("iloc")
print(df.iloc[1:5,3])
print("loc")
print(df.loc[1:5,"mpg"])
print("-----------------------------------------------------------------------------------------------------------------------")
print(df.describe(include="all"))
print("-----------------------------------------------------------------------------------------------------------------------")
print(df.describe(include="float"))
print("-----------------------------------------------------------------------------------------------------------------------")
print(df.describe(percentiles=[0.35,0.87]))
```

```python
#Correlation
# 1 ==> perfect ,, 0.8,0.9 ==> good ,, less than 0.8 ==> bad
df.corr()
#df.corr(numeric_only=True)  for laptop
```

```python
# graphicall form 
import matplotlib.pyplot as plt
pd.plotting.scatter_matrix(df,figsize=[25,25],marker="*",alpha=0.7)
plt.show()
```

```python
# Qualitative Data vs Quantitative Data
from pandas.plotting import parallel_coordinates
pll=parallel_coordinates(df,"cylinders",cols=["mpg","acceleration","origin"],color=["red","yellow","green","orange","blue"])
plt.legend()
plt.show()
```
