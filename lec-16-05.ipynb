```python
#13
age=[]
for i in df['Age']:
    if i<=18:
        age.append("child")
    elif i>18 and i<=30:
        age.append("adult")
    elif i>30 and i<=50:
        age.append("young")
    else:
        age.append("senior")
        
df['age_group']=age
pd.crosstab(df['Survived'],df['age_group'],rownames=['survived'],colnames=['age_group'])
```
